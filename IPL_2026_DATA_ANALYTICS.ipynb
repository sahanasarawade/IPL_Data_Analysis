```python
# dropna
df_demo = df.copy()
print("Shape before dropping rows:",df_demo.shape)
df_demo_dropped = df_demo.dropna()
print("Shape after dropping rows:",df_demo_dropped.shape)
print("\nWe would lose:",df_demo.shape[0] - df_demo_dropped.shape[0])
```

```python
# Drop constant columns

# First, let's confirm these columns really are constant (only 1 unique value)

constant_check_cols = [
    'Season',
    'Umpire_1',
    'Umpire_2',
    'Orange_Cap',
    'Purple_Cap',
    'Tournament_MVP',
    'Emerging_Player',
    'Fair_Play_Award',
    'Player_of_the_Series',
    'Final_Awards'
]

for c in constant_check_cols:
    print(f"{c}: {df[c].nunique()} unique value(s)")
```

```python
# All of them have only 1 unique value, so it is safe to drop them
columns_to_drop = [
    'Season',
    'Umpire_1',
    'Umpire_2',
    'Orange_Cap',
    'Purple_Cap',
    'Tournament_MVP',
    'Emerging_Player',
    'Fair_Play_Award',
    'Player_of_the_Series',
    'Final_Awards'
]

df = df.drop(columns=columns_to_drop,errors='ignore')

print("Columns dropped! New shape:", df.shape)
```

```python
# Add values in missing column using mean() -> attendance col
mean_attendance = df['Attendance'].mean()
print("Mean attendance:", round(mean_attendance, 2))

df['Attendance'] = df['Attendance'].fillna(mean_attendance)

print("Missing values left in Attendance:", df['Attendance'].isnull().sum())
```

```python
# add missing value with the help of median

for col in ['Powerplay_Runs', 'Home_Impact_Runs', 'Away_Impact_Runs']:
    median_value = df[col].median()
    df[col] = df[col].fillna(median_value)
    print(f"{col} -> filled with median value: {median_value}")

print("\nRemaining missing values:")
print(df[['Powerplay_Runs', 'Home_Impact_Runs', 'Away_Impact_Runs']].isnull().sum())
```

```python
# Filling value with the help of mode

mode_wides = df['Wides'].mode()[0]
print("Most common number of wides:", mode_wides)

df['Wides'] = df['Wides'].fillna(mode_wides)

print("Missing values left in Wides:", df['Wides'].isnull().sum())
```

```python
# Fill missing values in Text Column

df['Player_of_the_Match'] = df['Player_of_the_Match'].fillna('Not Available')

print("Missing values left in Player_of_the_Match:",
      df['Player_of_the_Match'].isnull().sum())

# Final check: Are there any missing values left anywhere in the dataset?

print("\nTotal missing values remaining in the whole dataset:",
      df.isnull().sum().sum())
```

```python
# Rename any column
# Rename any column

df = df.rename(columns={'Match_Duration_Minutes': 'Match_Duration'})

print("Column renamed successfully. Check:",
      'Match_Duration' in df.columns)
```

```python
# Change data type

df['Date'] = pd.to_datetime(df['Date'])

df['Attendance'] = df['Attendance'].astype(int)

df['Powerplay_Runs'] = df['Powerplay_Runs'].astype(int)

print(df[['Date', 'Attendance', 'Powerplay_Runs']].dtypes)
```

```python
# removing data from rows permanently

rows_before = df.shape[0]

df = df.drop_duplicates()
df = df.reset_index(drop=True)

rows_after = df.shape[0]

print(f"Removed {rows_before - rows_after} duplicate row(s).")
print("Final cleaned shape:", df.shape)
```

```python
# Home Team Wins Count
home_wins = df[df["Winner"] == df["Home_Team"]]["Home_Team"].value_counts()

# Away Team Wins Count
away_wins = df[df["Winner"] == df["Away_Team"]]["Away_Team"].value_counts()

teams = sorted(set(home_wins.index) | set(away_wins.index))

home = [home_wins.get(team, 0) for team in teams]
away = [away_wins.get(team, 0) for team in teams]

x = np.arange(len(teams))
w = 0.35

plt.figure(figsize=(12, 6))
plt.bar(x - w/2, home, width=w, label="Home Wins", color="green")
plt.bar(x + w/2, away, width=w, label="Away Wins", color="orange")

plt.xticks(x, teams, rotation=90)
plt.xlabel("Teams")
plt.ylabel("Number of Wins")
plt.title("Home vs Away Wins by Team")
plt.legend()
plt.tight_layout()
plt.show()
```
